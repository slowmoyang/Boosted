```python
import ROOT
from easy_jet import EasyJet
easy_jet = EasyJet()
```

```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from itertools import cycle
colors = cycle(mpl.colors.CSS4_COLORS.keys())
```

```python
root_file = ROOT.TFile.Open("/data/slowmoyang/Boosted/ppTOtttt-QCD.root", "READ")
tree = root_file.Delphes
tree.GetEntry(0)
```

```python
jets = easy_jet.cluster_from_tree(tree)
```

```python
fig, ax = plt.subplots(figsize=(9, 9))

for jet_idx in range(jets.size()):
    color = colors.next()
    for each in jets[jet_idx].constituents():
        ax.scatter(each.rap(), each.phi(), color=color, alpha=0.5)

axis_label_kwargs = {
    "fontdict": {"size": 20}
}
ax.set_xlabel("$\eta$", **axis_label_kwargs)
ax.set_ylabel("$\phi$", **axis_label_kwargs)
ax.grid()
```
