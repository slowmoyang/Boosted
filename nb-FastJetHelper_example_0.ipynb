```python
import ROOT
from fastjet_utils import FastJetHelper
fastjet_helper = FastJetHelper()
```

```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from itertools import cycle
colors = cycle(mpl.colors.CSS4_COLORS.keys())
```

```python
root_file = ROOT.TFile.Open("/data/slowmoyang/Boosted/ppTOtttt-QCD.root", "READ")
tree = root_file.Delphes
tree.GetEntry(0)
```

```python
jets = fastjet_helper.cluster_from_tree(tree)
```

```python
fig, ax = plt.subplots(figsize=(9, 9))

for jet_idx in range(jets.size()):
    color = colors.next()
    for each in jets[jet_idx].constituents():
        ax.scatter(each.rap(), each.phi(), color=color, alpha=0.5)

ax.set_xlabel("$\eta$")
ax.set_ylabel("$\phi$")
ax.grid()
```
